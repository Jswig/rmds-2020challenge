```python
import pandas as pd
import numpy as np
import requests
```

```python
url = "https://opendata.arcgis.com/datasets/9681f6a5269f4b6997b398bca057b62c_0.geojson"
r = requests.get(url)
data = r.json()
```

```python
county_cases = {}

for i, d in enumerate(data['features']):
    properties = d['properties']
    county_cases.update({properties['COMTY_NAME']:{'cases':properties['cases'], 'case_rate':properties['case_rate'], 'deaths':properties['deaths'], 'death_rate':properties['death_rate']}})
    
df = pd.DataFrame(county_cases).T
```
